### Implementación del Shooting Method ###

```python
def ode_solver(t, y0, dydt):
  y = np.zeros([len(y0), len(t)])
  y[:,0] = y0
  for i in range(1,len(t)):
    h = t[i]-t[i-1]
    y[:,i] = y[:,i-1] + h*dydt(t[i-1], y[:,i-1])
  return y

def ShootingMethod(s, P, N=20):
  """
  Implemenst the Shooting Method
  using the given initial slope s
  """
  # Define problem as dynamic system
  f = P["f(t,y,y')"]
  def ydot(t, y):
    y1, y2 = y
    return np.array([y2, f(t,y1,y2)])
  # Define initial condition using the given slope
  y0 = np.array([P["y(a)"], s])
  # Solve using a ODE solver
  t_num = np.linspace(P["a"], P["b"], N)  
  y_num = ode_solver(t_num, y0, ydot)
  y_num = y_num[0,:]
  # Return values
  return t_num, y_num
```

### Código para Graficar ###

```python
def plotSolution(P_sol, P):
  # Construct the solution
  N = 201 # For sampling the true solution
  t = np.linspace(P["a"], P["b"], N)
  y = P_sol["y(t)"]
  # Plot the figure on interactive mode
  fig = plt.figure(figsize=(12,8))
  plt.suptitle(P_sol["title"], fontsize=20)
  #plt.plot(t, y(t), "k", lw=2.0, label="$y^{(true)}(t)$", alpha=0.5)
  plt.grid('on')
  plt.xlabel("$t$", fontsize=20)
  plt.ylabel("$y(t)$", fontsize=20)
  return

def plotShootingMethod(s, P, N=20):
  # Get the solution
  t_num, y_num = ShootingMethod(s, P,N)
  # Add the information to the plot
  plt.plot(t_num, y_num, 'o:', label="$y_s(t)$, $s=%.2f$"%s)
  # Add to legend
  plt.legend(loc="best", numpoints=1, fontsize=20)
  return

def show():
  plt.show()
```

En realidad deseamos encontrar precisamente un $s$ donde $y_s(b)=y_b$, es decir, encontrar un cero de la función $F(s)=y_s(b)-y_b$.

La función $F(s)$ es unidimensional, por lo que podemos graficarla convenientemente.

```python
def F(s, P):
  t_num, y_num = ShootingMethod(s, P)
  return P["y(b)"] - y_num[-1]

def plotF(P, s_min=-2., s_max=4., Ns=10):
  s_list = np.linspace(s_min, s_max, Ns)
  F_list = [F(s,P) for s in s_list]
  fig = plt.figure(figsize=(12,8))
  plt.plot(s_list, F_list)
  plt.xlabel("$s$", fontsize=20)
  plt.ylabel("$F(s)=y_b-y_s(b)$", fontsize=20)
  plt.grid('on')
  plt.show()
```

Podemos encontrar $s$ tal que $F(s)=y_s(b)-y_b =0$ utilizando un algoritmo de bisección.

```python
def bisection(s_left, f_left, s_right, f_right, f, tol=1E-6):
  if f_left*f_right>0:
    print "Error, not valid backeting"
    return None
  s_middle = .5*(s_left + s_right)
  f_middle = f(s_middle)
  print "s=%+.8f, F(s)=%+.8f" %(s_middle, f_middle)
  if abs(f_middle)<tol:
    return s_middle
  elif f_middle*f_left<0:
    return bisection(s_left, f_left, s_middle, f_middle, f, tol)
  else:
    return bisection(s_middle, f_middle, s_right, f_right, f, tol)
```
